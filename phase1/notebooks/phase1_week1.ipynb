```python
def isiterable(obj):
    """ A function that checks if an object is iterable """
    try:
        iter(obj)
        return True
    except TypeError: # not iterable
        return False
```

```python
isiterable([1,2]) # True
isiterable("hello") # True
isiterable(5) # False
```

```python
def add_and_maybe_multiply(a, b, c= None):
    """ adds to numbers, if a third is provided multiply that sum by it"""
    result = a + b

    if c is not None:
        result = result * c

    return result

print(add_and_maybe_multiply(3, True))
print(add_and_maybe_multiply(3, 5, False))
```

```python
words = ["apple", "bat", "bar", "atom", "book"]
by_letter = {}
for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
print(by_letter)
```

```python
# better version of the above

by_letter = {}

for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)

print(by_letter)
```
