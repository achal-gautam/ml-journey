```python
def isiterable(obj):
    """ A function that checks if an object is iterable """
    try:
        iter(obj)
        return True
    except TypeError: # not iterable
        return False
```

```python
isiterable([1,2]) # True
isiterable("hello") # True
isiterable(5) # False
```
